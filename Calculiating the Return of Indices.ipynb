```python
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['^GSPC', '^IXIC', '^GDAXI']
```

```python
##          S&     NASDAQ   ÍNDICE ALEMÃO DAX   NASDAQ(FOOTSIE)
```

```python
ind_data = pd.DataFrame()
```

```python
for t in tickers:
    ind_data[t] = wb.DataReader(t, data_source='yahoo',start='1997-1-1')['Adj Close']
```

```python
ind_data.head()
```

```python
ind_data.tail()
```

```python
(ind_data / ind_data.iloc[0] * 100).plot(figsize=(14,8))
plt.show()
```

```python
ind_returns = (ind_data / ind_data.shift(1)) - 1
ind_returns.tail()
```

```python
anual_ind_returns = ind_returns.mean() * 250
anual_ind_returns
```
